```python
import pandas as pd
import json
```

```python
with open ("t20_wc_match_results.json") as f:
    data=json.load(f)
df=pd.DataFrame(data[0]["matchSummary"])

df.head()
```

```python
df.shape
```

```python
df.rename({"scorecard":"matchid"},axis=1,inplace=True)
df.head()
```

```python
df.to_csv("tempmatch.csv",index=False)
```

```python
my_match_dict={}

for index,row in df.iterrows():
    key1=row["team1"]+" Vs "+row["team2"]
    key2=row["team2"]+" Vs "+row["team1"]
    
    my_match_dict[key1]=row["matchid"]
    my_match_dict[key2]=row["matchid"]

my_match_dict
```

# batting Summary

```python
with open ("t20_wc_batting_summary.json") as f1:
    data= json.load(f1)
    all_records=[]
    
    for rec in data:
        all_records.extend(rec["battingSummary"])
        
df_bat=pd.DataFrame(all_records)

df_bat.head(10)
```

```python
df_bat["out/notout"]=df_bat.dismissal.apply(lambda x:"out" if len(x) > 0 else "not_out" )
```

```python
df_bat.head()
```

```python
df_bat.drop(columns=["dismissal"],inplace=True)
df_bat.head(12)
```

```python
df_bat["batsmanName"]=df_bat["batsmanName"].apply(lambda x:x.replace("â€",""))
```

```python
df_bat.head(12)
```

```python
df_bat["matchid"]=df_bat["match"].map(my_match_dict)
```

```python
df_bat.head()
```

```python
df_bat.to_csv("tempbat.csv",index=False)
```

# balling

```python
with open ("t20_wc_bowling_summary.json") as f2:
    data= json.load(f2)
    all_records=[]
    
    for rec in data:
        all_records.extend(rec["bowlingSummary"])
        
df_ball=pd.DataFrame(all_records)

df_ball.head(30)
```

```python
df_ball["matchid"]=df_ball["match"].map(my_match_dict)
```

```python
df_ball.head(10)
```

```python
df_ball.to_csv("tempball.csv",index=False)
```
